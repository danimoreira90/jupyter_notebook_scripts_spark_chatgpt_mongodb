```python
""" O Guia de estilo para código Python afirma que cada script deve começar com uma docstring que explica o propósito do script"""
""" Comparação de inteiros usando declaração ID e operadores. """

print('Informe dois números inteiros para', 
      'verificar a relação entre eles.')

# Primeiro número inteiro
number1 = int(input('Informe o primeiro número inteiro: '))

# Segundo número inteiro
number2 = int(input('Informe o segundo número inteiror: '))

if number1 == number2: # leia == como “é igual a” e = como “é atribuído”
    print(number1, 'é igual ao', number2)

if number1 != number2:
    print(number1, 'é diferente', number2)

if number1 < number2:
    print(number1, 'é menor que', number2)

if number1 > number2:
    print(number1, 'é maior que', number2)

if number1 <= number2:
    print(number1, 'é menor que ou igual à', number2)

if number1 >= number2:
    print(number1, 'é maior que ou igual a', number2)
```

# Tipagem dinâmica

```python
"""Python usa tipagem dinâmica — ela determina o tipo de objeto ao qual uma variável se refere durante a execução do seu código."""
x = 4
print(type(x))
x = 4.1
print(type(x))
x = 'Aula'
print(type(x))
# Python cria objetos na memória e os remove da memória conforme necessário.
# A medida que o valor da variável altera, o dado anterior é liberado da memória - Esse processo — chamado coleta de lixo
```

```python
"""Encontrar o mínimo de três valores"""
number1 = int(input('Primeiro número: '))
number2 = int(input('Segundo número: '))
number3 = int(input('Terceiro número: '))

minimum = number1  

if number2 < minimum:
    minimum = number2

if number3 < minimum:
    minimum = number3

print('O menor valor é:', minimum)
```

```python
print('O menor valor é:',min(number1,number2,number3))
```

# Cálculo IMC
IMC = peso / (altura x altura)

```python
#IMC
nome = input('Informe o seu nome: ') 
altura = float(input('Altura: ')) #1.85
peso = float(input('Peso: ')) #115
imc = peso / altura ** 2

print(nome, 'tem', altura, 'de altura,', 'pesa',peso,'quilos')
print('Seu IMC é', imc)
```

```python
# Formatando IMC
print("O valor do IMC formatado é: {:.2f}".format(imc))
```

```python
# Arredondamento
print("O valor do IMC formatado é: "+str(round(imc, 2)))
```

```python
"f-strings"
linha_1 = f'{nome} tem {altura:.2f} de altura,'
linha_2 = f'pesa {peso} quilos e seu imc é'
linha_3 = f'{imc:.2f}'

print(linha_1)
print(linha_2)
print(linha_3)
```
